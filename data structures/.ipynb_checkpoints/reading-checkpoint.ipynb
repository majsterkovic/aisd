```python
import matplotlib.pyplot as plt

res = open('cmake-build-debug/res.txt', 'r')

fig = plt.figure()
fig.set_size_inches(20, 16)

data = res.readlines()
print(data)
methods = ["push", "remove"]
types = ["random", "sorted", "rev_sorted", "A", "V"]
structures = ["linkedlist", "bst"]

d = {}
for m in methods:
    d[m] = {}
    for s in structures:
        d[m][s] = {}
        for t in types:
            d[m][s][t] = []

n = 0
m = 0

for line in data:
    struct, method, typ, size, time = line.split(';')
    time = int(time)
    d[method][struct][typ].append(time)

res.close()

print(d)
```
