```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import matplotlib
from tinydb import TinyDB, Query

res = open('cmake-build-debug/res.txt', 'r')

fig = plt.figure()
fig.set_size_inches(20, 16)

data = res.readlines()

sizes = ["100", "500", "1000", "5000", "10000", "50000", "100000"]
methods = ["push", "remove"]
types = ["random", "sorted", "rev_sorted", "A", "V"]
structures = ["linkedlist", "bst"]

db = TinyDB('db.json')

for line in data:
    struct, method, typ, size, time = line.split(';')
    time = int(time)
    db.insert({'struktura' : struct,
              'metoda' : method,
              'typdanych' : typ,
              'wielkosc' : size,
              'czas' : time})

res.close()



```

```python
X = Query()
lista = db.search((X.struktura == "linkedlist") & (X.metoda == "push") & (X.wielkosc == "100000") & (X.typdanych == "random"))
print(lista)
```
